```python
!pip install fredapi
from fredapi import Fred
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
```

```python
start_date = '2022-01-01'
end_date = '2023-07-27'
ticker = 'GOOGL MSFT TSLA AMZN NVDA GOOG META TSM'

stock_performance_df = yf.download(ticker, start=start_date, end=end_date, period='1mo')['Close']

fig = go.Figure()

for column in stock_performance_df.columns:
    fig.add_trace(go.Scatter(x=stock_performance_df.index, y=stock_performance_df[column], mode='lines', name=column))

fig.update_layout(
    title={
        'text': '<b>Monthly Stock Performance of Tech Giants: Jan 2022 - Jul 2023</b>'.format(start_date, end_date),
        'x': 0.5,
        'y': 0.95,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': dict(size=18, color='white')
    },
    yaxis_title='Closing Price (USD)',
    xaxis=dict(type='date'),
    yaxis=dict(tickprefix='$'),
    legend_title='Ticker',
    template='plotly_dark',
    width=850,
    height=500,
)

fig.show()
```
